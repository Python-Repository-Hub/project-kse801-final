```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
loss = np.load('../output/gsi/loss.npy')
x = range(len(loss))
with plt.style.context('bmh'):
    font = {'color': 'darkred', 'size': 12, 'family': 'serif'}
    font_legend = {'size': 12, 'family': 'serif'}

    fig, axs = plt.subplots(1, 1, figsize=(8, 5))

    axs.plot(x, loss, label='Loss', color='tomato', linewidth=2, alpha=1)
    axs.set_xlabel('Epoch', fontdict=font)
    axs.set_ylabel('Loss', fontdict=font)
    axs.set_title('Loss on Gaussian Square Init Dataset', fontdict=font)
    axs.legend(loc='upper right', prop=font_legend)
    labels = axs.get_xticklabels() + axs.get_yticklabels()
    [label.set_fontname('serif') for label in labels]
plt.savefig('../output/gsi/loss.png')
```
